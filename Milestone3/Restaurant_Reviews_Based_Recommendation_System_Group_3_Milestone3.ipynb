**C) Research Question:**

Our goal here is to predict each consumer's ratings for the restaurants. A list of restaurants with the highest predicted ratings can be recommended to each consumer. Because the restaurant ratings are numerical, predicting their values can be a regression problem.

 A list of restaurants with the highest predicted ratings can be recommended to each consumer. Because the restaurant ratings are numerical, predicting their values can be a regression problem. Using classification techniques is also reasonable since this dataset's possible ratings are three discrete integers. However, a three-class classification will likely predict too many ties and prevent us from generating a top list of recommendations for a consumer. Rather than predicting the exact values of ratings that a consumer would give certain restaurants, what is more, important is predicting the ranks of these restaurants for the consumer. Thus, I will predict continuous values for consumer ratings. 

 I will use linear models to predict the ratings. The fundamental concepts are that a vector of features characterizes each restaurant; each consumer preference describes a vector of weights with the same dimension as the product features, and the predicted rating of a restaurant-consumer pair equals the inner product of the two vectors. The mathematics is, therefore, equivalent to a simple linear regression. For a content-based approach, the features are already determined based on the product information, leaving the weights to be optimized. Instead of using an existing library for recommender systems, I will implement the cost functions of the problem.

<h3> k-Nearest Neighbors (k-NN) Regression Model </h3>

```python
from sklearn import neighbors
from sklearn.metrics import mean_squared_error as mse
import math

val_rmse = []
for k in range(30):
    k = k+ 1
    knn1 = neighbors.KNeighborsRegressor(n_neighbors = k, weights='uniform', 
                                         algorithm='auto')

    knn1.fit(train_X_scale,train_y)
    pred = knn1.predict(test_X_scale) 
    rmse_error = math.sqrt(mse(test_y,pred))
    val_rmse.append(rmse_error)
    print('The RMSE value for k=',k,'is',rmse_error)
```

```python
# Function to evaluate ROC score for a Multi-Class Model
from sklearn.preprocessing import LabelBinarizer

def multiclass_roc_auc_score(test_y, pred_y, average="macro"):
    lb = LabelBinarizer()
    lb.fit(test_y)
    test_y = lb.transform(test_y)
    pred_y = lb.transform(pred_y)
    return roc_auc_score(test_y, pred_y, average=average)
```

<h3>Few Additional Observations</h3>

 We adjust the restaurant features the users see. Most of the elements in this matrix are one, except for those corresponding to 'parking_lot,' 'smoking_area,' 'alcohol,' and 'price': 
 
 (1) If a user doesn't have a car, they wouldn't care if there is parking, and the corresponding weight for 'parking_lot' is not needed. 
 
 (2) If a user doesn't smoke, they wouldn't care if there is a smoking area, and a corresponding weight for 'smoking_area' is not needed. 
 
 (3) Users with different 'drink_level' care differently about whether alcohol is provided in the restaurant.
 
 (4) Users with different 'budgets care differently about the price. People with higher budget cares less.
